#### Q6. Explain the concept of generators in Python and how they are defined.
#### Ans:
#### Generator:  A generator is a special type of iterable that allows you to create an iterator using a function that contains one or more yield statements. Unlike regular functions that return a single value and terminate, a generator can yield multiple values, one at a time, while maintaining its state between each yield. This makes generators memory-efficient and ideal for processing large datasets or infinite sequences.

```python
#for example:
def fib(n):
    a=0
    b=1
    for i in range(n):
        yield a
        a, b = b, a+b
f = fib(10000000)
print(f)
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))
```

#### Q7. What are the advantages of using generators over regular functions?
#### Ans:
#### 1) Memory Efficiency: Generators yield items one at a time, allowing you to iterate through large data sets without loading everything into memory. This is particularly useful for processing large files or streams.
#### 2) Lazy Evaluation: Generators compute values on-the-fly as needed, which can lead to performance improvements by avoiding unnecessary calculations.
#### 3) Simplified Code: Using yield can make the code cleaner and easier to read compared to using a traditional function with return statements and state management.
#### 4) State Retention: Generators maintain their state between calls, making it easier to manage complex iterations without having to explicitly track the state.
#### 5) Pipelining: Generators can be composed together to create pipelines of operations, allowing for a more modular and flexible approach to data processing.
#### 6) Concurrency: Generators can be used to manage concurrent tasks with less overhead, as they can yield control without the complexity of threading.
#### 7) These advantages make generators a powerful tool in Python for various programming tasks, especially when dealing with potentially large or infinite sequences.





```python
#for example
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

# Using the generator
for num in fibonacci(10):
    print(num)
```

#### Q10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
#### Ans: 
#### 1. map():
#### Purpose: Transforms each element in a sequence (e.g., list, tuple) by applying a function to it.
#### Usage: map(function, iterable)
#### Output: Returns a map object, which is an iterator with each element processed by the function. Often converted to a list or tuple.

```python
#for example map()
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```

#### 2. filter():
#### Purpose: Filters elements in a sequence based on whether they satisfy a given condition.
#### Usage: filter(function, iterable)
#### Output: Returns a filter object, an iterator with only the elements that satisfy the condition.

```python
#for example filter() function
only_Odd = filter(lambda x: x if x%2==1 else "", [1,2,3,4,5,6,7,8,9])
print(list(only_Odd))
```

#### 3. reduce():
#### Purpose: Reduces a sequence into a single cumulative value by repeatedly applying a function to pairs of elements.
#### Usage: reduce(function, iterable) (requires importing from functools)
#### Output: A single value after all elements are reduced to one result.

```python
#for example reduce() function
from functools import reduce    
from math import gcd
def lc(a,b):
    return abs(a*b) // gcd(a,b) # gcd of a and b
lcm = reduce(lc, [1,2,3,4,5,6,7])
print(lcm)
```

#### Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number ** 0.5) + 1): # Check divisibility from 2 up to the square root of number
        if number % i == 0:
            return False
    return True
    
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number")
```

#### Q7. Implement a generator function that reads a file line by line and yields each line as a string.


```python
def read_file(file_path):
    """Generator function that reads a file line by line."""
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Yield each line without leading/trailing whitespace

# Example usage:
file_path = '/home/piyushkurwade/Desktop/Git Course/three/one.txt' 
for line in read_file(file_path):
    print(line)  # Process each line (for example, print it)
```
